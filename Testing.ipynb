```python
# https://ioos.github.io/notebooks_demos/notebooks/2017-12-15-finding_HFRadar_currents/
def get_csw_records(csw, filter_list, pagesize=10, maxrecords=1000):
    """Iterate `maxrecords`/`pagesize` times until the requested value in
    `maxrecords` is reached.
    """
    from owslib.fes import SortBy, SortProperty

    # Iterate over sorted results.
    sortby = SortBy([SortProperty("dc:title", "ASC")])
    csw_records = {}
    startposition = 0
    nextrecord = getattr(csw, "results", 1)
    while nextrecord != 0:
        csw.getrecords2(
#            constraints=filter_list,
            startposition=startposition,
            maxrecords=pagesize,
            sortby=sortby,
        )
        csw_records.update(csw.records)
        if csw.results["nextrecord"] == 0:
            break
        startposition += pagesize + 1  # Last one is included.
        if startposition >= maxrecords:
            break
    csw.records.update(csw_records)
```

```python
maxrecords = csw.results['matches']
print(maxrecords)
filter_list = None
get_csw_records(csw, filter_list, pagesize=10, maxrecords=maxrecords)
```

```python
user_agent_url = 'https://fbinter.stadt-berlin.de/fb/wms/senstadt/ortsteil/?service=wms&request=GetCapabilities'
xml_data = requests.get(user_agent_url).text
```

```python
parser = et.XMLParser(encoding="utf-8")
tree = et.fromstring(xml_data, parser=parser)
```

```python
print(list(tree))
```

```python
for child in tree:
    print('1 ==>',''.join(child.itertext()), child.text)
    for node in child:
        print('2 ==>',''.join(node.itertext()), node.text)
        for elem in node:
            print('3 ==>',''.join(elem.itertext()), elem.text)
            for tag in elem:
                print('4 ==>',''.join(tag.itertext()), tag.text)
                for li in tag:
                    print('5 ==>',''.join(li.itertext()), li.text)
```
